```python
v= list(map(int, [i for i in np.arange(0.1,21,1)]))
coss= np.cos(v)
exp= np.exp(v)
fig, axes= plt.subplots(1,2, figsize=(8,3), dpi=100)
axes[0].plot(v, coss)
axes[1].plot(v, exp)
```

```python
# or on the same axis:

v= list(map(int, [i for i in np.arange(0.1, 21, 1)]))
cos= np.cos(v)
exp= np.exp(v)
fig, axes= plt.subplots(1,1, figsize=(8,3), dpi=100)
axes.plot(v,cos, color= 'r')
axes.set_ylabel('cos values')
axes.set_xlabel('numbers')

ax1= axes.twinx()
axes.set_ylabel('exponential values')
ax1.plot(v,exp, color= 'b')

plt.show()
```
