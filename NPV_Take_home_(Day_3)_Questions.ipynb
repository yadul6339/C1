```python
df
```

```python
f= df['Sales'].groupby([df['Store'], df['Location']])
print(f.min().unstack())
print(f.max().unstack())
```

```python
#another way of doing it 

import numpy as np
f= df.pivot_table(values= 'Sales', index= 'Store', columns= 'Location', aggfunc=[np.min, np.max])
```
